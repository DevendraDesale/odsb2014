```python
pacific_type_a = %sql select TO_DATE('01/01/'||TO_CHAR(year),'MM/DD/YYYY')+(week*7) as surveyed_on, a_total from us_who_flu_stats where region='Pacific' order by year asc, week asc
```

```python
pacific_a = pacific_type_a.DataFrame()
pacific_a.set_index('surveyed_on')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(16,6));
pacific_a.plot(x='surveyed_on', y='a_total', ax=axes, title='total type A infections', label='infections')
```
