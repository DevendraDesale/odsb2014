Previously, we looked at how to do batch-based copies of data from a database into HDFS, the Hadoop filesystem.  Big Data SQL, is somewhat different: it allows the database to declare a table on data in HDFS.  It can then be queried without copying or loss of performance.

Big Data SQL requires an external table, so let's declare on of type `ORACLE_HDFS` on our `flu_search` data in HDFS.

```python
%%sql CREATE TABLE flu_search (search_doc VARCHAR2(4000))
  ORGANIZATION EXTERNAL (
  TYPE ORACLE_HDFS
  DEFAULT DIRECTORY DEFAULT_DIR
   LOCATION ('/user/oracle/flu_search/part*')
  ) 
  REJECT LIMIT UNLIMITED
```

The data we stored in HDFS was JSON formatted, so for now we can keep the table as having a single column.

```python
%sql select * from flu_search where rownum < 10
```

Even if we want to keep the table definition as a single column, a [database view](http://en.wikipedia.org/wiki/View_%28SQL%29) allows us to treat the JSON as if it were a table with columns.

```python
%%sql create or replace view flu_view as
  select
    cast (f.search_doc.term as varchar2(255)) term,
    cast (f.search_doc.score as number) score,
    to_date(f.search_doc.month, 'DD-MM-YYYY') term_date
  from flu_search f
```

Now we've got something that maps nicely to a DataFrame!

```python
%sql select * from flu_view where rownum < 10
```

Before that, though, let's take our plotting code and wrap it up in a method.  This will make it easier to plot the results of any SQL query on our trend data.

```python
def plot_term_search(term_search):
    term_search = term_search.DataFrame()
    term_search.index = pd.DatetimeIndex(term_search['term_date'])
    term_search['score'] = term_search['score'].astype(float)
    gt = term_search.groupby("term")
    fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(16,12));
    for t,g in gt:
        g.score.plot(label=t)
    plt.legend(loc="best")
```

A quick spot check using terms we know should be associated with H5N1 validates our thinking.  There were large spikes in the news around "flu" and "outbreak" at the same time as "h5n1."  Of course, these terms are likely from the same news stories.

```python
s = %sql select * from flu_view where term in ('flu', 'h5n1', 'outbreak') order by term_date
plot_term_search(s)
```

Let's look again at the the h5n1 incident data, but explicitly at the years 2006, and 2007.  If we group by both country and year, we can determine where the most reported incidents were.  The `extract` sql function will pull the year out of our `incident_date` column.

It appears that a set of European countries (UK, Turkey, Romania) and a set of Asian countries (Thailand, China, Vietnam) contributed to the total number of incidents in those years.

```python
%%sql select incident_location, extract(year from incident_date) as incident_year, 
    count(incident_date) as incidents from pathogenic_flu
    where extract(year from incident_date) in (2006,2007)
    group by incident_location, extract(year from incident_date) 
    order by incidents desc, incident_location, incident_year
    
```

Because most of the terms in our flu search data have been "stemmed," it may be difficult to find exact matches.  Instead we'll use the SQL `like` operator to get near matches for all of the countries of interest.  By placing these `like` comparison in a `case` statement, we can control the groups returned for plotting.

```python
%%sql select term_date, term, score from ( 
        select term_date, 
            case
                when term like 'british%' then 'british'
                when term like 'europe%' then 'europe'
                when term like 'viet%' then 'vietnam'
                when term like 'asia%' then 'asia'
                when term like 'turk%' then 'turkey'
                when term like 'ghana%' then 'ghana'
                when term like 'hong%' then 'hong kong'
                when term like 'asia%' then 'asia'
                when term like 'ukraine%' then 'ukraine'
                when term = 'h5n1' then 'h5n1'
                else null
            end as term,
            score
        from flu_view
    ) where term is not null
    order by term_date
```

Looking at the plot, it appears that turkey, vietnam, and ghana saw the most correlation in score with the h5n1 outbreaks of 2006 and 2007.  With this information we could quickly go back to our news data and cross check to see how many stories contained both h5n1 and one of the countries with highest score.  It's likely that news of h5n1 incidents increased the monthly relevance of these countries in our search system.
