# Do Vaccination Rates Impact Flu Rates?

Flu shots are important, as any doctor will tell you.  But can we use our data and statistics to explore how flu shots impact the flu?

In this exercise, we'll examine US flu and flu shot data to see if there are any meaningful conclusions.  Along the way, we'll use analytical SQL to find the year-over-year change in certain data.  We'll also build some simple [Ordinary Least Squares](http://en.wikipedia.org/wiki/Ordinary_least_squares) regression models to evaluate the significance of our assumptions.

To begin, start with a typical set of `import`s and the addition of the `statsmodel` Python library.

```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from __future__ import print_function
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.tools.tools import add_constant
%matplotlib inline
```

## Getting the Data

We'll begin by looking at our flu shot data.  The flu shot data is by state and the US flu surveillance data is by region, so we'll have to join with the `state_stats` table to map states to regions.  From there, we can determine a few more things:

* the number of eligible Medicare recipients in the region
* the average percentage of Medicare recipients to receive a flu shot in the region
* the year of the observation

```python
%%sql select sum(eligible) as all_eligible, avg(percentage_claimed) as vaccinated, region_name, year 
  from (
    select f.state, f.eligible, f.percentage_claimed, s.region_name, f.year
      from flushots f, state_stats s 
      where s.state=f.state and f.ethnicity='T' 
      and f.week=52 and year in (2012, 2013)
  ) group by year, region_name
```

In order to determine the impact of flu shots on the flu, we need to see if something changed.  That "something" is a *response variable*.  In our case, the response variable is the year-to-year change in type-A flu cases.

Our data only records totals, not the changes between them.  Fortunately, many databases include a `lag` function which allows us to compare a record to the one prior to it.  Let's use a SQL `lag` function to find year-over-year change in type-A flu.

```python
%%sql select total_a - lag(total_a) over (partition by region order by year), region, year
      from (
        select sum(a_total) as total_a, sum(b) as total_b, region, year
        from us_who_flu_stats 
        where year in (2012,2013,2014) group by year, region
       )
```

We could merge these two DataFrames together in python, but we can get a single DataFrame to work by writing a SQL clause which joins the results.  Notice that each of our previous queries is now a subquery.

```python
%%sql with shots as (
  select sum(eligible) as all_eligible, avg(percentage_claimed) as vaccinated, region_name, year 
  from (
    select f.state, f.eligible, f.percentage_claimed, s.region_name, f.year
      from flushots f, state_stats s 
      where s.state=f.state and f.ethnicity='T' 
      and f.week=52 and year in (2012, 2013, 2014)
  ) group by year, region_name
),
sick as (
  select total_a - lag(total_a) over (partition by region order by year) as yoy_change, region, year
  from (
    select sum(a_total) as total_a, region, year
    from us_who_flu_stats 
    where year in (2012, 2013,2014) group by year, region
    )
  )
select sick.yoy_change, sick.region, sick.year as sick_year, 
shots.year as shot_year, shots.all_eligible, shots.vaccinated
from sick, shots where sick.region=shots.region_name
and sick.year = (shots.year+1)
```

## Analyzing the Data

Now that we've got data to analyze, we should start by taking a look at it.  One of the best ways to do this is simply to scatter plot the data and look for patterns.  Look at the plot below: is there an obvious pattern between % of medicare patients who recieved flu shots and the yearly change in flu cases?

```python
shots_to_sick = _.DataFrame()
shots_to_sick['yoy_change'].fillna(0,inplace=True)
shots_to_sick['vaccinated'] = shots_to_sick['vaccinated'].astype(float)
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(16,12));
shots_to_sick.plot(kind="scatter", x='vaccinated', y='yoy_change',ax=ax)
```

Perhaps the number of people eligible has a more obvious pattern than the % of people vaccinated.  Let's consider another scatter plot.

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(16,12));
shots_to_sick.plot(kind="scatter", x='all_eligible', y='yoy_change',ax=ax)
```

### Building models

Both *factors* -- eligiblity and percentage vaccinated -- may have an effect on the response, but not visually apparent.  We can build a linear model and see if it explains the data well.  First we'll use the `ols` method in `statsmodels` to build a model that effectively states:

*Year-over-year change in flu is caused by the total number of eligible medicare recipients, the percentage of recipients vaccinated, and the percentage vaccinated multipled by the number of eligible.*

While this model is not likely accurate, the diagnostics on the model can give us insight into whether the factors explain any of the data. 

```python
shots_to_sick = add_constant(shots_to_sick)
shot_to_sick_model = ols("yoy_change ~ all_eligible*vaccinated", data=shots_to_sick).fit()
```

The diagnostic information contains a few interesting pieces:

* R-squared is only 0.309, which suggests that the model only explains 30% of the data
* No factors have a *p-value* (P>|t|) of less than or equal to 0.05, which suggests that none of them has a significant effect

Basically, our model is wrong.  This means we should ask another question: do we have the right factors to explain the data?

```python
print(shot_to_sick_model.summary())
```

### Another Model

Suppose we consider a different factor.  Perhaps the year-over-year change in flu cases is explained by access to flu shots, not strictly the increase in vaccinations.  In this case, the factor of interest would be the change in eligibility.  

Let's construct a new dataset to examine that scenario.

```python
%%sql with shots as (
  select all_eligible- lag(all_eligible) over (partition by region_name order by year) as eligible_change, region_name, year 
  from ( 
    select sum(f.eligible) as all_eligible, s.region_name, f.year
      from flushots f, state_stats s 
      where s.state=f.state and f.ethnicity='T' 
      and f.week=52 and year in (2012, 2013, 2014)
      group by year, region_name
  ) 
),
sick as (
  select total_a - lag(total_a) over (partition by region order by year) as yoy_change, region, year
  from (
    select sum(a_total) as total_a, region, year
    from us_who_flu_stats 
    where year in (2012, 2013,2014) group by year, region
    )
  )
select sick.yoy_change, sick.region, sick.year as sick_year, 
shots.year as shot_year, shots.eligible_change
from sick, shots where sick.region=shots.region_name
and sick.year = (shots.year+1)
and yoy_change is not null
```

As before, a quick scatter plot helps us understand what's going on in the data.  Note that there's a stronger linear relationship here -- you can almost draw a straight line through the dots.

```python
shots_to_sick = _.DataFrame()
shots_to_sick["eligible_change"].fillna(0,inplace=True)
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(16,12));
shots_to_sick.plot(kind="scatter", x='eligible_change', y='yoy_change',ax=ax)
```

Now we'll build another least squares model, this time the model states:

* Year-over-year change in flu is caused by the year-over-year change in eligible medicare recipients

As we see in the summary, the R-squared values suggests that we're explaining more of the data with fewer parameters.
Our new factor is also significant, as it has a *p-value* of less than 0.05.  It suggests we're on the right track.

```python
shots_to_sick = add_constant(shots_to_sick)
shot_to_sick_model = ols("yoy_change ~ eligible_change", data=shots_to_sick).fit()
print(shot_to_sick_model.summary())
```

One way to check the goodness of fit for a model is to consider a plot of the *residuals*.  Residuals are the difference between the value predicted by the model and the actual value observed in the dataset.  Large residuals suggest a model which is incorrect in some way.

In this case, we'll look at a "component plus residual plot."  The blue line gives us the predicted values, while the dots are the predicted values plus the residual (i.e., the original value).  Note that the majority of the dots lie near the line.

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(16,12));
sm.graphics.plot_ccpr(shot_to_sick_model, "eligible_change", ax=ax)
```

Overall, the model is better and we're on the right track.  However, change in Medicare eligibility isn't the only factor in reducing flu nationwide.  Could change in percentage vaccinating year-over-year be significant?  Are there other factors to consider?
