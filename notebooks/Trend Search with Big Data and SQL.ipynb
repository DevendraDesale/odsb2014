Previously, we looked at how to do batch-based copies of data from a database into HDFS, the Hadoop filesystem.  Big Data SQL, is somewhat different: it allows the database to declare a table on data in HDFS.  It can then be queried without copying or loss of performance.

Big Data SQL requires an external table, so let's declare on of type `ORACLE_HDFS` on our `flu_search` data in HDFS.

```python
%%sql CREATE TABLE flu_search (search_doc VARCHAR2(4000))
  ORGANIZATION EXTERNAL (
  TYPE ORACLE_HDFS
  DEFAULT DIRECTORY DEFAULT_DIR
   LOCATION ('/user/oracle/flu_search/part*')
  ) 
  REJECT LIMIT UNLIMITED
```

The data we stored in HDFS was JSON formatted, so for now we can keep the table as having a single column.

```python
%sql select * from flu_search where rownum < 10
```

Even if we want to keep the table definition as a single column, a [database view](http://en.wikipedia.org/wiki/View_%28SQL%29) allows us to treat the JSON as if it were a table with columns.

```python
%%sql create or replace view flu_view as
  select
    cast (f.search_doc.term as varchar2(255)) term,
    cast (f.search_doc.score as number) score,
    to_date(f.search_doc.month, 'DD-MM-YYYY') term_date
  from flu_search f
```

Now we've got something that maps nicely to a DataFrame!

```python
%sql select * from flu_view where rownum < 10
```

H5N1 is the clear standout here, with 2 really big peaks -- one between 2006 and 2007, and the other just in recent years.  Since there doesn't seem to be a correlated rise with the countries which had the most outbreaks, perhaps we need to consider different comparison criteria.

Before that, though, let's take our plotting code and wrap it up in a method.  This will make it easier to plot the results of any SQL query on our trend data.

```python
def plot_term_search(term_search):
    term_search = term_search.DataFrame()
    term_search.index = pd.DatetimeIndex(term_search['term_date'])
    term_search['score'] = term_search['score'].astype(float)
    gt = term_search.groupby("term")
    fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(16,12));
    for t,g in gt:
        g.score.plot(label=t)
    plt.legend(loc="best")
```

A quick spot check using terms we know should be associated with H5N1 validates our thinking.  There were large spikes in the news around "flu" and "outbreak" at the same time as "h5n1."  Of course, these terms are likely from the same news stories.

```python
s = %sql select * from flu_view where term in ('flu', 'h5n1', 'outbreak') order by term_date
plot_term_search(s)
```
