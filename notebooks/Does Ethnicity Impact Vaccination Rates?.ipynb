```python
vac_rate = %sql select ethnicity, percentage_claimed from flushots where week=52 and ethnicity in ('W', 'A', 'H', 'B')
vac_frame = vac_rate.DataFrame()
vac_frame
```

```python
vac_frame['percentage_claimed'] = vac_frame['percentage_claimed'].astype("float")
```

```python
vac_rates_grouped_by_eth = vac_frame.groupby('ethnicity').percentage_claimed
vac_rates_grouped_by_eth.describe()
```

```python
groups = [e for e in vac_rates_grouped_by_eth]

def pairwise_t_test(groups, sig=0.05):
    for i in range(len(groups)):
        grp = groups[i]
        for j in range(i,len(groups)):
            other = groups[j]
            if grp[0] != other[0]:
                r = ttest_ind(grp[1], other[1])
                reject = False
                if r[1] < sig:
                    reject=True
                yield grp[0], other[0], r[0], r[1], reject

t_test_results = pd.DataFrame(pairwise_t_test(groups), columns=["Group 1", "Group 2", "t stat.", "p-value", "reject H0"])
t_test_results
```
