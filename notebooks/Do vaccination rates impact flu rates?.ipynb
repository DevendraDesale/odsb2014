```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from __future__ import print_function
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.tools.tools import add_constant
%matplotlib inline
```

```python
%load_ext sql
%sql oracle://fludb:flushot@localhost:1521/orcl
```

```python
%%sql select sum(eligible) as all_eligible, avg(percentage_claimed) as vaccinated, region_name, year 
  from (
    select f.state, f.eligible, f.percentage_claimed, s.region_name, f.year
      from flushots f, state_stats s 
      where s.state=f.state and f.ethnicity='T' 
      and f.week=52 and year in (2012, 2013)
  ) group by year, region_name
```

```python
%%sql select total_a - lag(total_a) over (order by year), region, year
      from (
        select sum(a_total) as total_a, sum(b) as total_b, region, year
        from us_who_flu_stats 
        where year in (2012,2013,2014) group by year, region
       )
```

```python
%%sql with shots as (
  select sum(eligible) as all_eligible, avg(percentage_claimed) as vaccinated, region_name, year 
  from (
    select f.state, f.eligible, f.percentage_claimed, s.region_name, f.year
      from flushots f, state_stats s 
      where s.state=f.state and f.ethnicity='T' 
      and f.week=52 and year in (2012, 2013)
  ) group by year, region_name
),
sick as (
  select total_a - lag(total_a) over (order by year) as yoy_change, region, year
  from (
    select sum(a_total) as total_a, region, year
    from us_who_flu_stats 
    where year in (2013,2014) group by year, region
    )
  )
select sick.yoy_change, sick.region, sick.year as sick_year, 
shots.year as shot_year, shots.all_eligible, shots.vaccinated
from sick, shots where sick.region=shots.region_name
and sick.year = (shots.year+1)
```

```python
shots_to_sick = _.DataFrame()
shots_to_sick['vaccinated'] = shots_to_sick['vaccinated'].astype(float)
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(16,12));
shots_to_sick.plot(kind="scatter", x='vaccinated', y='yoy_change',ax=ax)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(16,12));
shots_to_sick.plot(kind="scatter", x='all_eligible', y='yoy_change',ax=ax)
```

```python
shots_to_sick = add_constant(shots_to_sick)
shot_to_sick_model = ols("yoy_change ~ all_eligible*vaccinated", data=shots_to_sick).fit()
```

```python
print(shot_to_sick_model.summary())
```

```python
%%sql with shots as (
  select all_eligible- lag(all_eligible) over (order by year) as eligible_change, region_name, year 
  from ( 
    select sum(f.eligible) as all_eligible, s.region_name, f.year
      from flushots f, state_stats s 
      where s.state=f.state and f.ethnicity='T' 
      and f.week=52 and year in (2012, 2013)
      group by year, region_name
  ) 
),
sick as (
  select total_a - lag(total_a) over (order by year) as yoy_change, region, year
  from (
    select sum(a_total) as total_a, region, year
    from us_who_flu_stats 
    where year in (2013,2014) group by year, region
    )
  )
select sick.yoy_change, sick.region, sick.year as sick_year, 
shots.year as shot_year, shots.eligible_change
from sick, shots where sick.region=shots.region_name
and sick.year = (shots.year+1)
and yoy_change is not null
```

```python
shots_to_sick = _.DataFrame()
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(16,12));
shots_to_sick.plot(kind="scatter", x='eligible_change', y='yoy_change',ax=ax)
```

```python
shots_to_sick = add_constant(shots_to_sick)
shot_to_sick_model = ols("yoy_change ~ eligible_change", data=shots_to_sick).fit()
print(shot_to_sick_model.summary())
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(16,12));
sm.graphics.plot_ccpr(shot_to_sick_model, "eligible_change", ax=ax)
```
