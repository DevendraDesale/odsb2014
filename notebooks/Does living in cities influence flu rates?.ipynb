```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.graphics.api import interaction_plot, abline_plot
from statsmodels.stats.anova import anova_lm
%matplotlib inline
```

```python
%%sql select p.urban_pop, p.rural_pop, sum(f.TOTAL_A_INFECTIONS/p.total_pop) as per_cap_a, 
sum(f.total_b_infections/p.total_pop) as per_cap_b, f.country 
from flu_statistics f, population_info p
where f.country=p.country
and f.year=p.year
group by f.country, p.urban_pop, p.rural_pop
order by urban_pop
```

```python
urban_flu = _.DataFrame()
for c in ["urban_pop", "rural_pop", "per_cap_a", "per_cap_b"]:
    urban_flu[c] = urban_flu[c].astype(float)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(16,12));
urban_flu.plot(kind="scatter", x="urban_pop", y="per_cap_a", ax=ax)
ax.set_ylim(-0.0005,0.002)
```

```python
urban_model = ols("per_cap_a ~ urban_pop", data=urban_flu).fit()
print(urban_model.summary())
```

```python
sm.graphics.plot_regress_exog(urban_model, "urban_pop", fig)
```

```python
%%sql select g.percapgdp as per_cap_gdp, p.urban_pop, 
sum(f.TOTAL_A_INFECTIONS/p.total_pop) as per_cap_a, f.country 
from flu_statistics f, population_info p, gdp_data g
where f.country=p.country
and g.country=f.country
and f.year=p.year
and g.year=f.year
group by f.country, g.percapgdp, p.urban_pop
```

```python
urban_gdp = _.DataFrame()
for c in ['per_cap_gdp', 'urban_pop', 'per_cap_a']:
    urban_gdp[c] = urban_gdp[c].astype(float)
```

```python
urban_gdp_model = ols("per_cap_a ~ per_cap_gdp*urban_pop", data=urban_gdp).fit()
print urban_gdp_model.summary()
```

```python
def pop_quantiles(p):
    if p < 25:
        return "very low"
    elif p >= 25 and p < 50:
        return "low"
    elif p >= 50 and p < 75:
        return "moderate"
    elif p >= 75 and p < 100:
        return "high"
    else:
        return "very low"
urban_gdp['pop_q'] = pd.Series(map(pop_quantiles, urban_gdp['urban_pop']))
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(16,12));
f=interaction_plot(urban_gdp['per_cap_gdp'], urban_gdp['pop_q'], urban_gdp['per_cap_a'], ms=10, ax=ax)
```
