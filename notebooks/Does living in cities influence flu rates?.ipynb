# Does Living in Cities Influence Flu Rates?

Along with GDP and other factors, the amount of a country's population which is either urban or rural may impact flu transmission rates.  For example, do more densely-packed countries with high urban population percentages transmit the disease more easily than largely agrarian countries?

In this exercise, we'll examine the effect of rural and urban population on flu rates using basic [linear regression](http://en.wikipedia.org/wiki/Linear_regression) modeling.  Specifically, we'll use [ordinary least squares](http://en.wikipedia.org/wiki/Ordinary_least_squares) to quickly model the impacts of these factors on a response variable: per capita flu rates. 

As usual, we import the necessary libraries and connect to our database.

```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.graphics.api import interaction_plot, abline_plot
from statsmodels.stats.anova import anova_lm
from statsmodels.tools.tools import add_constant
%matplotlib inline
```

## Population

As with GDP and other factors, we really want to focus on how flu rates are impacted *per capita*.  We don't want factors like total population to cloud our ability to analyze the data.  As such, we'll need to extract the data from the database in a fashion that gives us per capita rates for type-A and type-B flu infection.  This is very simple: we sum the total infections for each country and divide them by the population of each country.

```python
%%sql select p.urban_pop, p.rural_pop, sum(f.TOTAL_A_INFECTIONS/p.total_pop) as per_cap_a, 
sum(f.total_b_infections/p.total_pop) as per_cap_b, f.country 
from flu_statistics f, population_info p
where f.country=p.country
and f.year=p.year
group by f.country, p.urban_pop, p.rural_pop
order by urban_pop
```

For ease of use, we'll put the result in a DataFrame and convert the decimal fields to `float` types for numpy and statsmodels.

```python
urban_flu = _.DataFrame()
for c in ["urban_pop", "rural_pop", "per_cap_a", "per_cap_b"]:
    urban_flu[c] = urban_flu[c].astype(float)
```

As with almost all of our analyses, we want to look at the data before we make modeling decisions.  The scatterplot is our best friend in this case as we can look quickly to see if there are obvious relationships in the data.  Here we look at urban population on the x-axis and per capita type-A infections on the y-axis.  Can you see a trend?  Is it clear?

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(16,12));
urban_flu.plot(kind="scatter", x="urban_pop", y="per_cap_a", ax=ax)
ax.set_ylim(-0.0005,0.002)
```

There appears to be at least some indication that increased urban populations are associated with higher per capita flu rates.  Let's build a linear model and see if the relationship is significant.

```python
urban_flu = add_constant(urban_flu)
urban_model = ols("per_cap_a ~ urban_pop", data=urban_flu).fit()
print(urban_model.summary())
```

Looking at the results table, we can see that the *p-value* (P>|t|) for urban_population is significant, and does impact type-A rates.  However, R-squared and adjusted R-squared suggest that urban population percentage does not fully explain the data.

A look at the residuals can show us where our model predicts poorly.  `statsmodels` includes the `plot_regress_exog` function, which will give us a set of residual plots.  Notice that as urban population percentage increase, the model explains fewer of the data points.

```python
sm.graphics.plot_regress_exog(urban_model, "urban_pop", fig)
```

## Population and GDP

Perhaps as countries become more urban, their GDP impacts transmission.  It could be that urban cities which are poor have high transmission rates.  To examine this, let's extract a similar dataset, but join in data about GDP.

```python
%%sql select g.percapgdp as per_cap_gdp, p.urban_pop, 
sum(f.TOTAL_A_INFECTIONS/p.total_pop) as per_cap_a, f.country 
from flu_statistics f, population_info p, gdp_data g
where f.country=p.country
and g.country=f.country
and f.year=p.year
and g.year=f.year
group by f.country, g.percapgdp, p.urban_pop
```

As before, we store the results in a DataFrame and convert decimal types to floats.

```python
urban_gdp = _.DataFrame()
for c in ['per_cap_gdp', 'urban_pop', 'per_cap_a']:
    urban_gdp[c] = urban_gdp[c].astype(float)
```

We're going to construct another linear model, but this time we're going to use 2 factors: urban population and per capita GDP.  We're only going to consider the *interaction* between the two factors.  There may be situations in which GDP only has a strong effect when urban population is high, or vice versa.

```python
urban_gdp = add_constant(urban_gdp)
urban_gdp_model = ols("per_cap_a ~ per_cap_gdp:urban_pop", data=urban_gdp).fit()
print urban_gdp_model.summary()
```

The results in this case seem significant and explain almost 20% more of the data.  It's not a perfect fit, but it is indicative of an interaction effect.

We can now check the residuals and see that we do have more of the actual values close to our model's equation.

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(16,12));
sm.graphics.plot_regress_exog(urban_gdp_model, "per_cap_gdp:urban_pop", fig);
```

We can also look at an *interaction plot* to determine the magnitude of interaction at various levels.  Let's break the urban population percentage into quartiles and add it as a new column in our DataFrame.

```python
def pop_quantiles(p):
    if p < 25:
        return "very low"
    elif p >= 25 and p < 50:
        return "low"
    elif p >= 50 and p < 75:
        return "moderate"
    elif p >= 75 and p < 100:
        return "high"
    else:
        return "very low"
urban_gdp['pop_q'] = pd.Series(map(pop_quantiles, urban_gdp['urban_pop']))
```

The interaction plot doesn't show us a perfect, linear interaction between GDP and urban population percentage, but we can see that there's a much more pronounced effect from high GDP when the population is highly urban.

We cannot claim to have fully explained the data, but we can say that there is a likely interaction between these factors on per capita type-A rates. 

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(16,12));
f=interaction_plot(urban_gdp['per_cap_gdp'], urban_gdp['pop_q'], urban_gdp['per_cap_a'], ms=10, ax=ax)
```
