# How Do Sanitation and Clean Water Effect Flu Transmission?

The conditions in which we live have profound effects on our health.  Just as the percentage of a country's population which is urban may impact flu transmission, sanitation and availability of clean water may play a role in how flu is spread.

In this exercise, we'll examine the effect of improvements in sanitation and potable water accessibility improvement on the flu.  We'll do this by examining groups of data in parallel, modeling the most likely candidate, and exploring the concept of [*logistic regression*](http://en.wikipedia.org/wiki/Logistic_regression).

To begin, we'll need our standard set of `import`s and a new one from [scikit-learn](http://scikit-learn.org/stable/).

```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.graphics.api import interaction_plot, abline_plot
from statsmodels.stats.anova import anova_lm
from statsmodels.tools.tools import add_constant
from sklearn import linear_model, datasets
%matplotlib inline
```

First, we'll write a simple SQL query to give us per capita infections of each flu type as well as each country's year-over-year improvement in sanitation and water access.  Note that the World Bank breaks water improvements into rural and urban, so we'll want both columns.  We can also grab data for cellular phone access and primary education.

```python
%%sql select s.region, sum(s.total_a_infections/p.total_pop) as per_cap_a, sum(s.total_b_infections/p.total_pop) as per_cap_b,
  d.country, d.sanitation, d.rh2o, d.uh2o, d.cellular, d.primary_ed
  from flu_statistics s, dev_indc d, population_info p
    where d.country=s.country
    and p.country=d.country
    and d.sanitation is not null
    and d.rh2o is not null
    and d.uh2o is not null
    and d.cellular is not null
    and d.primary_ed is not null
    group by d.country, d.sanitation, d.rh2o, d.uh2o, d.cellular, d.primary_ed, s.region
```

In typical fashion, we put the data into a Pandas DataFrame and convert decimal types to float.

```python
devel_frame = _.DataFrame()
for c in ["per_cap_a", "per_cap_b", "sanitation", "uh2o","rh2o","cellular", "primary_ed"]:
    devel_frame[c] = devel_frame[c].astype(float)
```

In order to determine if sanitation plays a role on flu transmission, we should first look to see if there are notable differences in sanitation across the WHO regions.  We can do this by plotting a simple histogram of each region.

```python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(16,12));
devel_frame['sanitation'].hist(by=devel_frame['region'], ax=ax, color='crimson')
```

The regions do appear to be very different, as such, we should not model them together.  Let's group each region and compute some correlations between our factors of interest and per capita flu rates.

```python
regions = devel_frame.groupby(by="region")
```

```python
models = {}
for name, group in regions:
    correlations = {}
    correlations["sanitation"] = pearsonr(group['sanitation'], group['per_cap_a'])
    correlations["uh2o"] = pearsonr(group['uh2o'], group['per_cap_a'])
    correlations["rh2o"] = pearsonr(group['rh2o'], group['per_cap_a'])
    models[name] = correlations
models
```

Looking at the correlations above, it appears that the strongest correlations are in the AFRO region.  Moreover, the correlation with urban improvements in water supply are less pronounced than in the rural case.  As such, let's attempt to model the AFRO region by itself.

We'll need to create a new DataFrame that only includes the rows in which AFRO is the region.  Unfortunately, we only have 15 points of data.  It will be hard to get any sort of significant fit.  Regardless, let's fit a model that considers the interactions between sanitation and water improvement.

```python
afro_rows = devel_frame['region'] == 'AFRO'
afro_data = devel_frame[afro_rows]
print afro_data.describe()
afro_data = add_constant(afro_data)
afro_model = ols("per_cap_a ~ sanitation:rh2o + sanitation:rh2o:uh2o", data=afro_data).fit()
print afro_model.summary()
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(16,12));
sm.graphics.plot_partregress_grid(afro_model, fig=fig)
```

While the effects are statistically significant, the relationship explained in the data -- and more pronouncedly in the residual plots -- suggests that the relationship is not linear.  Instead, most of our residual points are clustered around a single area, and there are a few outliers.  With more data, the model may be more meaningful, but we can't conclude anything with so little data.

However, looking at the data, it could be that rather than a linear effect on the data, there's a *binary* one.  That is, perhaps sanitation and water supply improvements divide AFRO countries into 2 groups.  One way we can examine this is using a logistic regression.

We still don't have enough data to form a significant model, but we can construct a logistic regression model to understand the concept.

We're going to be using logistic regression in a supervised fashion.  That is, we need to provide the "right" answers for a set of data which will *train* our model.  We can then use other data to *test* our model.  This training-testing cycle is typical in machine learning applications.

Let's start by hand-classifying AFRO countries into 2 groups, 0 and 1.

```python
afro_data["class"] = 0
afro_data.at[afro_data["per_cap_a"] <= 0.000007, "class"] = 0
afro_data.at[afro_data["per_cap_a"] > 0.000007,"class"] = 1
afro_data
```

From here, we can divide the DataFrame into training and testing sets.  Given the small amount of data, we'll devote most of it to training, ideally validation sets should be at least approximately the same size as training sets.  We'll select our training set rows at random, choosing 12 values.

```python
print len(afro_data.index.values)
training_set_index = np.random.choice(afro_data.index.values, 12, replace=False)
in_training = lambda x: x not in training_set_index and x or None
testing_set_index = np.array(filter(lambda x: x != None, [in_training(i) for i in afro_data.index.values]))
training_set = afro_data.ix[training_set_index]
testing_set = afro_data.ix[testing_set_index]
training_set
```

From here, building a logistic regression model is simple.  The `sklearn.linear_model.LogisticRegression` object takes training parameters for tolerance, iterations and other features of the algorithm.  We then call `fit` with `X` the training data factors and `y` the training data responses.

```python
X = training_set[["sanitation", "rh2o"]]
y = training_set["class"]
lr = linear_model.LogisticRegression(C=1e5, tol=0.000001)
lr.fit(X,y)
```

There are many ways to examine logistic regression outputs, but scatterplots work well.  First, we'll examine the *actual* classification of the training set.

```python
plt.prism()
plt.scatter(X.sanitation, X.rh2o, c=y)
```

Now we'll look at where the testing set would appear in the plot.  Can you guess what class they should belong to?

```python
plt.prism()
plt.scatter(testing_set["sanitation"], testing_set["rh2o"], c='white', marker='^')
```

Now let's look at the predictions from our model.  Plotting the predicted data as triangles shows us their location and classification.  The `score` function of our logistic regression object allows us to evaluate it's predictive performance.

In our initial run, we managed to correctly predict the outcomes.  However, because the training set is chosen randomly, subseqeuent runs of this experiment may produce different levels of accuracy.  Indeed, with such a small number of observations, even a change of one row may have huge impacts on the model produced!

```python
y_pred_train = lr.predict(X)
plt.scatter(X["sanitation"], X["rh2o"], c=y_pred_train)
testing_y = lr.predict(testing_set[["sanitation", "rh2o"]])
plt.scatter(testing_set["sanitation"], testing_set["rh2o"], c=testing_y, marker="^")
print "Accuracy on test set:", lr.score(testing_set[["sanitation","rh2o"]], testing_set["class"])
training_outcome = pd.DataFrame()
training_outcome["predicted"] = y_pred_train
training_outcome["actual"] = np.array(y)


testing_outcome = pd.DataFrame()
testing_outcome["predicted"] = testing_y
testing_outcome["actual"] = np.array(testing_set["class"])
```
