```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.tools.tools import add_constant
%matplotlib inline
```

```python
%load_ext sql
%sql oracle://fludb:flushot@localhost:1521/orcl
```

```python
%%sql select g.percapgdp as per_cap_gdp, sum(f.TOTAL_A_INFECTIONS/p.total_pop) as per_cap_a, 
sum(f.total_b_infections/p.total_pop) as per_cap_b, f.country 
from flu_statistics f, population_info p, gdp_data g
where f.country=p.country
and g.country=f.country
and f.year=p.year
and g.year=f.year
and g.percapgdp is not null
group by f.country, g.percapgdp
```

```python
per_cap_data = _.DataFrame()
for c in ['per_cap_gdp', 'per_cap_a', 'per_cap_b']:
    per_cap_data[c] = per_cap_data[c].astype(float)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(16,12));
per_cap_data.plot(kind="scatter", x="per_cap_gdp", y="per_cap_a", ax=ax)
```

```python
pearsonr(per_cap_data['per_cap_gdp'], per_cap_data['per_cap_a'])
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(16,12));
per_cap_data.plot(kind="scatter", x="per_cap_gdp", y="per_cap_b", ax=ax)
```

```python
pearsonr(per_cap_data['per_cap_gdp'], per_cap_data['per_cap_b'])
```

```python
per_cap_data = add_constant(per_cap_data)
gdp_model = ols("per_cap_a ~ per_cap_gdp", data=per_cap_data).fit()
print(gdp_model.summary())
```

```python
sm.graphics.plot_ccpr(gdp_model, "per_cap_gdp", ax=ax)
```
