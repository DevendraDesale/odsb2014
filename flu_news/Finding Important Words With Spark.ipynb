## Determining Word Importance: Stemming and TF-IDF

First, we're going to revisit the question of word-importance.  A previous lesson illustrated that the most frequently used words in English are typically not what news stories are *about*.  Let's load our news data as an RDD and examine a story.  It's obvious to us what the story is about, but the number of small, connecting words dwarfs those we care about.

For speed, we'll start by taking a smple of our news data.  This is done with the built-in RDD function `sample`.

```python
import json
import re
from stemming.porter2 import stem
news_rdd = sc.textFile("hdfs://localhost:8020/user/oracle/flu_news")
news_json = news_rdd.sample(False, 0.1,2).map(lambda x: json.loads(x))
print news_json.take(5)[2]
```

To really measure word importance, we're going to need to clean up and normalize the text.  This involves a few steps, which we put into the `clean_word` function.

+ Make all words lowercase
+ Remove special characters and punctuation
+ *Stem* the words -- i.e., "talk" and "talking" should both be treated as "talk"

Stemming is an important part of working with text.  While there are many packages which accomplish this, we'll use a simple python package to simplify things.

With this function in place, we can create two RDDs which may be interesting: the full text and the titles.  We'll be focusing on the full text for obvious reasons.  We're going to create a special RDD in this case: a *PairRDD*.  This allows us to identify the full text of an article by its title: it's referred to as a *key-value* pair.  The key is the title and the value is the full text.

In order to perform our TF-IDF transformation, we'll also need the total frequency of each word in the data set.  This is equivalent to the word-count we performed previously.  However, we'll use Spark's `countByKey` method to provide us a Python dict to work with.

```python
def clean_word(w):
    return re.sub("\,|\.|\;|\:|\;|\?|\!|\[|\]|\}|\{", "", stem(w.lower()))
document_bodies = news_json.map(lambda x: (x['title'], " ".join(map(lambda y: clean_word(y), x['fulltext'].split()))))
document_titles = news_json.map(lambda x: x['title'])
all_terms = document_bodies.map(lambda x: (x[0], list(set(x[1].split()))))
term_document_count = all_terms.flatMap(lambda x: [(i, x[0]) for i in x[1]]).countByKey()
term_document_count.items()[:5]
```

Once we have the PairRDD for our document bodies and the dictionary of all terms, we can start to look at the TF part of TF-IDF.  TF stands for "term frequency."  Given our new expertise in counting words, this should be simple to calculate.  Again, we'll use a Spark built-in `countByValue` to produce a dictionary.

```python
document_tf = document_bodies.map(lambda x: (x[0], x[1].split())).flatMapValues(lambda x: x).countByValue()
document_tf.items()[:5]
```

We'll also need some other information for the computation of IDF, or *inverse document frequency*.  That is, we want to scale term frequency (how often a word appears) by how common that word is across all documents.  The more unique that word is to the article, the more important it may be.

To do this, we need to compute a few more things:
+ the total number of documents
+ the total number of words in each document

Spark's counting functions make this easy.

```python
import numpy as np
total_documents = 1.0 * document_bodies.count()
total_words_per_article = document_bodies.map(lambda x: (x[0], x[1].split())).flatMapValues(lambda x: x).countByKey()
total_words_per_article.items()[:5]
```

There's one last thing we need to make our TF-IDF computation: how common each word is, or how many articles does the word appear in.  To do this, we'll start by making an RDD of distinct words per article.  You'll notice this is like our `term_document_count` from earlier.

```python

distinct_words_per_article = document_bodies.map(lambda x: (x[0], x[1].split())).flatMapValues(lambda x: x).distinct()

word_occurences_across_articles = distinct_words_per_article.map(lambda x: (x[1], x[0])).countByKey()
word_occurences_across_articles.items()[:5]
```

Once this is done, we can define a function to compute TF-IDF for each document given a set of dictionaries.  The function `article_tf_idf` does just this.

```python
def article_tf_idf(article_total, words_per_article, tf_per_article, occ_across_articles):
    result = []
    for key, value in tf_per_article.items():
        article = key[0]
        term = key[1]
        wpm = words_per_article[article]
        ocm = occ_across_articles[term]
        tf_idf = (float(value)/wpm) * np.log(article_total/ocm)
        result.append({"article":article, "term":term, "score":tf_idf})
    return result
```

Applying the function to our dictionary yields the sort of results we'd expect.  Words like "live" and "cup" have higher scores than words like "than" and "have."  We've also put the results into a list of dictionaries, which makes integration with pandas and matplotlib simple.

```python
article_word_importance = article_tf_idf(total_documents, total_words_per_article, document_tf, word_occurences_across_articles)
article_word_importance[:10]
```

## Examining Word Importance Over Time

Our TF-IDF function allows us to determine how important words are in each article, which gives us a better idea of what the articles are about.  However, what if we want to understand how important a word is to the news over time?  Is flu more important in March, 2010 than in April 2011?  Can we use TF-IDF to determine that?

In fact, we can, it just requires reframing the problem.  Instead of considering words inside of articles, we want to consider words inside of date ranges.  If we do that, then our approach can be applied easily.

Let's get started by making a function `date_to_key` which will help us construct PairRDDs for each month,year combination in the news dataset.  Then we can get an RDD `documents_over_time` which has dates for keys and article text for values.

```python
def date_to_key(d):
    return tuple(map(lambda x: int(x), d.split("-")[0:2]))
```

```python
documents_over_time = news_json.map(lambda x: (date_to_key(x['date']), x['fulltext'])).map(lambda x: (x[0], map(clean_word, x[1].split()))).map(lambda x: (x[0], " ".join(x[1])))
```

At this point, we have a number of values for each key.  That's going to make computing our TF-IDF over time more difficult.  What we really want is a bag of words for all article text for a given month.  We can use a reducer to do this.  We'll just append all the articles for the month together with spaces.

```python
wordbags_over_time = documents_over_time.reduceByKey(lambda x, y: x + " " + y)
```

From here, computing TF is just like doing it for an article!

```python
tf_per_month = wordbags_over_time.map(lambda x: (x[0], x[1].split())).flatMapValues(lambda x: x).countByValue()
tf_per_month.items()[0]
```

Similarly, we can compute the dictionaries we need for the rest of our TF-IDF computation just like before.

```python
total_words_per_month = wordbags_over_time.map(lambda x: (x[0], x[1].split())).flatMapValues(lambda x: x).countByKey()
total_words_per_month.items()[0]
```

```python
total_months = wordbags_over_time.count()
total_months
```

```python
distinct_words_per_month = wordbags_over_time.map(lambda x: (x[0], x[1].split())).flatMapValues(lambda x: x).distinct()
word_occurences_across_months = distinct_words_per_month.map(lambda x: (x[1], x[0])).countByKey()
word_occurences_across_months.items()[:5]
```

Finally, we make a new function for dealing with our monthly data.  The only change is in the dictionary keys of the result.

```python
def monthly_tf_idf(month_total, words_per_month, tf_per_month, occ_across_months):
    result = []
    for key, value in tf_per_month.items():
        month = key[0]
        term = key[1]
        wpm = words_per_month[month]
        ocm = occ_across_months[term]
        tf_idf = (float(value)/wpm) * np.log(month_total/ocm)
        result.append({"month":month, "term":term, "score":tf_idf})
    return result
```

```python
search_results = monthly_tf_idf(total_months, total_words_per_month, tf_per_month, word_occurences_across_months)
```

We can also take the resulting dictionary back into Spark for more transformations.  This is done by calling `sc.parallelize` on any list

```python
search_results_rdd = sc.parallelize(search_results)
```

We've finally got some monthly results to work with, so let's define some helper functions and import pandas.

```python
import pandas as pd

def to_ascii(v):
    try:
        return str(v)
    except:
        return ""

def flatten_for_data_frame(r):
    y,m= r['month']
    return {"date":pd.datetime(y,m,1),"term":r['term'],"score":r['score']}
    
def flatten_for_write(r):
    r['term'] = to_ascii(r['term'])
    return r
```

We'll also import our standard visualization modules so that we can plot the results.

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

Using our parallelized results, let's use spark's filter function to make a dataframe that only contains "flu" and a few other terms which might have high importance over time.  We can use `collect` to bring the results directly into a DataFrame.

```python
search_results_df = pd.DataFrame(search_results_rdd.map(flatten_for_data_frame).filter(lambda x: x['term'] in ("flu", "influenza", "rhinovirus", "cold", "china", "palestine")).collect())
search_results_df[:10]
```

When we plot the results, it become apparent that something's missing.  Even though we can see spikes in relative importance, we're missing lots of dates.  It makes some sense: not every day has a news story about flu or china.

```python
print "articles to compare", search_results_rdd.count()
grouped_terms = search_results_df.sort_index(by=["date"]).groupby("term") 
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(16,12));
for term, grp in grouped_terms:
    plt.plot(grp['score'], label=term)
plt.legend(loc="best")
```

To fix this, we're going to need to change the index of our DataFrame.  This is a bit complicated but once we have an index
that includes all the dates, we can swap our original DataFrame index for the new one.  At this point, the plot begins to look a bit better!

```python
min_date = min(search_results_df['date'])
max_date = max(search_results_df['date'])
mft2 = pd.DataFrame(index=pd.DatetimeIndex(pd.date_range(min_date,max_date,freq="MS")), columns=["date"])
search_history = pd.merge(mft2, search_results_df, left_index=True, right_on="date", how="left")
search_history.index=search_history.date
search_history.index
gt = search_history.groupby("term")
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(16,12));
for t,g in gt:
    g.score.plot(label=t)
plt.legend(loc="best")
```

Looking at the plot, there's still one thing missing: *data*!  Recall that we've been working with a sample of the original data.  Let's apply the same steps to a new set of RDDs derived from all the news data and look at the result.

```python
all_news = news_rdd.map(lambda x: json.loads(x))

all_documents_over_time = all_news.map(lambda x: (date_to_key(x['date']), x['fulltext'])).map(lambda x: (x[0], map(clean_word, x[1].split()))).map(lambda x: (x[0], " ".join(x[1])))

all_wordbags_over_time = all_documents_over_time.reduceByKey(lambda x, y: x + " " + y)

complete_tf_per_month = all_wordbags_over_time.map(lambda x: (x[0], x[1].split())).flatMapValues(lambda x: x).countByValue()

complete_total_words_per_month = all_wordbags_over_time.map(lambda x: (x[0], x[1].split())).flatMapValues(lambda x: x).countByKey()

complete_total_months = all_wordbags_over_time.count()

complete_distinct_words_per_month = all_wordbags_over_time.map(lambda x: (x[0], x[1].split())).flatMapValues(lambda x: x).distinct()

complete_word_occurences_across_months = complete_distinct_words_per_month.map(lambda x: (x[1], x[0])).countByKey()

complete_search_results = monthly_tf_idf(complete_total_months, complete_total_words_per_month, complete_tf_per_month, complete_word_occurences_across_months)
```

```python
complete_search_results_rdd = sc.parallelize(complete_search_results)
complete_search_results_df = pd.DataFrame(complete_search_results_rdd.map(flatten_for_data_frame).filter(lambda x: x['term'] in ("flu", "influenza", "rhinovirus", "cold", "china", "palestine")).collect())
```

```python
min_date = min(complete_search_results_df['date'])
max_date = max(complete_search_results_df['date'])
mft2 = pd.DataFrame(index=pd.DatetimeIndex(pd.date_range(min_date,max_date,freq="MS")), columns=["date"])
complete_search_history = pd.merge(mft2, complete_search_results_df, left_index=True, right_on="date", how="left")
complete_search_history.index=complete_search_history.date

gt = complete_search_history.groupby("term")
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(16,12));
for t,g in gt:
    g.score.plot(label=t)
plt.legend(loc="best")
```

Looking at the above plot, we can see real trends.  Can you identify world events which might have caused spikes in news importance?  Having this sort of data available for all the words in the news would allow people to find all sorts of interesting answers.  As such, we should save our results in a way that others can consume.

We'll use Spark's `saveAsTextFile` method to store our complete TF-IDF results in Hadoop's filesystem.  Later, we'll return to that data to build an interactive notebook using SQL and Hadoop.

```python
complete_search_results_rdd.map(flatten_for_write).filter(lambda x: x['term'] != "").saveAsTextFile("hdfs://localhost:8020/user/oracle/flu_search")
```
